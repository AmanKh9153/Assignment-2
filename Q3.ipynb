#### Step 7: (5 marks) Write your ideas of how to improve the spell checker. You don’t have to implement these ideas.

As we can see from the previous `bad_word` list, or the words, which are not present in the dictonary, Most of the words are simply rejected due to following reasons.
* When we are breaking the review using space, the punctuation marks remain attached with some words, which makes a mismatch. This happens mostly with the words, which appeared at the end of the sentenses ans carrying a trailing dot (.).<br>**Solution:** As a solution to this problem, we need to replace all possible special characters with space before breaking the review into words.
* Next problem is with the case of the letters in words. Here, our matching is case-sensetive, so if there is a mismatch in case of any letter in word it flags as mismatch. The words appearing at the beginning of any sentense suffers most from this problem, and the `words_list` obtained from '`nltk` also contains words with mixed cases.<br>**Solution:** To resolve this issue, it is a common practice to convert all words of dictonary to lower cases and then append into bloom filter, and before any processing on the reviews, we need to convert the whole review in lower case.
* Now, one small problem is, we tend to use numbers like, some year, date, time, money as rent amount, phone number etc., which is outside of dectionary and plays no role in spell checking. **Solution:** Replace occurances of all digits with spaces in the review.
* If any person writes his review in his local language, or in a language outside of our dictonary, then it will be treated as new words.**Solution:** We should introduce words from multiple languages in our dictonary.

Implementation of solution of some of the problems mentioned above,

```python
lower_word_filter = BloomFilter(max_elements=236736)

for word in word_list:
  lower_word_filter.add(word.lower())
```

```python
def string_cleaner(s):
  s = s.lower()
  sp_char = ['`', '~', '!', '@', '#', '$', '%', '^', '&', '*',
             '(', ')', '_', '-', '+', '=', '{', '}', '[', ']',
             '\\', '|', ';', ':', '\'', '"', ',', '<', '.', '>',
             '/', '?', '1', '2', '3', '4', '5', '6', '7', '8', '9', '0']
  
  for c in sp_char:
      while c in s: s = s.replace(c, ' ')
  return s
```

```python
lower_bad_words = dict()

for text in all_reviews:
  text = string_cleaner(text)
  for word in text.split():
    if word not in word_filter:
      if word in lower_bad_words: lower_bad_words[word] += 1
      else: lower_bad_words[word] = 1

len(lower_bad_words)
```

We have tried our best to scrap the popular websites, but we failed, because almost all of them are loading data dynamically using JavaScript, and BeautifulSoup is not capable to do so, we must have to use bworser automation. So to keep the thing simple, we choose a particular hotel from Mumbai containing more than 1000 reviews, and we are collecting all reviews of that particular Hotel. https://www.goibibo.com/hotels/ginger-mumbai-andheri-hotel-in-mumbai-999694052223892382/

```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
reviews = []

for index in range(1, 26):
  link = 'https://www.goibibo.com/hotels/ginger-mumbai-andheri-hotel-in-mumbai-999694052223892382/?page='+str(index)
  page = bs(req.get(link).text, "html.parser")
  spans = page.find_all('span', class_='UserReviewstyles__UserReviewTextStyle-sc-1y05l7z-4 cHVFbF')
  for span in spans:
    if not span.find_all('span'): reviews.append(span.text)

print("Total", len(reviews), "Reviews collected")
```

```python
reviews
```

```python
new_bad_words = dict()

for text in reviews:
  text = string_cleaner(text)
  for word in text.split():
    if word not in word_filter:
      if word in new_bad_words: new_bad_words[word] += 1
      else: new_bad_words[word] = 1

len(new_bad_words)
```
